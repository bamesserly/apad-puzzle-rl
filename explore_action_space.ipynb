```python
%load_ext autoreload
%autoreload 2
import numpy as np
from apad_env import APADEnv
import matplotlib.pyplot as plt
```

```python
def plot_action_space(action_space_size, logy = False):
    x = range(1,len(action_space_size)+1)
    if logy:
        plt.yscale('log')
    plt.scatter(x, action_space_size)
    plt.show()
```

# April 14

```python
action_list = []

env = APADEnv(4,14)

action_list.append(env.encode_action(7,0,0,0))
action_list.append(env.encode_action(1,0,0,2))
action_list.append(env.encode_action(2,0,1,32))
action_list.append(env.encode_action(0,1,0,12))
action_list.append(env.encode_action(6,0,0,4))
action_list.append(env.encode_action(3,0,0,7))
action_list.append(env.encode_action(4,0,0,28))
action_list.append(env.encode_action(5,0,0,30))

obs,info = env.reset()
obs_list = []
episode_starts = []
first_step = True
action_space_size = []
for action in action_list:
    obs, reward, terminated, truncated, info = env.step(action)
    obs_list.append(obs)
    mask = info["action_mask"]
    action_space_size.append(np.flatnonzero(mask).size)
    episode_starts.append(first_step)
    first_step = False

plot_action_space(action_space_size, True)
print(action_space_size)
env.visualize()
```

```python
from apad_env import APADEnv
from backtracking import solve
```

```python
# Solve a specific date
actions = solve(month=4, day=14, verbose=True)
print(f"Solution: {len(actions)} moves" if actions else "No solution")

# Visualize
if actions:
    env = APADEnv(4, 14)
    env.reset()
    for action in actions:
        env.step(action)
    env.visualize()

# Benchmark
# stats = benchmark_dates(n_samples=10, verbose=True)
# print(f"\nAvg solve time: {stats['mean_time_ms']:.1f}ms")
```
